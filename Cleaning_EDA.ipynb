Algerian Forest Fires Dataset Attributes:

1. Day: The day of the month when the observation was recorded.
2. Month: The month when the observation was recorded.
3. Year: The year when the observation was recorded.
4. Temperature: The temperature in degrees Celsius, which can influence fire behavior.
5. RH: Relative Humidity, expressed as a percentage, which affects fire intensity.
6. Ws: Wind Speed, in km/h, which impacts fire spread.
7. Rain: Amount of rainfall, in mm, which can affect the likelihood of fires.
8. FFMC: Fine Fuel Moisture Code, a fire weather index indicating moisture content in fine fuels.
9. DMC: Duff Moisture Code, a fire weather index for moisture content in deeper, decomposed fuels.
10. DC: Drought Code, a fire weather index indicating moisture levels in larger fuels.
11. ISI: Initial Spread Index, a fire weather index indicating potential fire spread rate.
12. BUI: Build-Up Index, a fire weather index representing the total amount of fuel available for combustion.
13. FWI: Fire Weather Index, a composite index providing an overall measure of fire danger.
14. Classes: Classification of the fire (e.g., whether a fire occurred or not), often used as the target variable in predictive modeling.


```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the Dataset
df=pd.read_csv("Algerian.csv",header=1)
```

```python
#head of the dataste
df.head()
```

```python
#information of the dataset
df.info()
```

Data Cleaning

```python
#missing value
df[df.isnull().any(axis=1)]
```

Replace Bejaia Region & Sidi-Bel Abbes Region Dataset with new Region column and values 0 for Bejaia Region , 1 for Sidi-Bel Abbes Region.

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
df.info()
```

```python
#change the type of Region column from float to int
df[['Region']]=df[['Region']].astype('int16')
df.info()
```

```python
#again check null values
df.isnull().sum()
```

```python
#drop null values
df=df.dropna().reset_index(drop=True)
```

```python
#after dropping null values agrain check null values to identify it.
df.isnull().sum()
```

```python
df.head()
```

```python
#now check the 122nd row
df.loc[[122]]
```

```python
#122nd row has again fault it has columns names so we drop it.
df=df.drop(122).reset_index(drop=True)
df.loc[[122]]
```

```python
#columns
df.columns
```

```python
#fix spaces in column name
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
#change the required columns as integer datatype.
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
df.info()
```

```python
#changing the columns into float which are required.
df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']]=df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype(float)
df.info()
```

```python
#for more information about dataset 
df.describe()
```

```python
#lets save the cleaned dataset.
df.to_csv('Algerian_clean',index=False)
```
